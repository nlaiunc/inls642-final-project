## Cleaning the MVPs dataset

```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.preprocessing import StandardScaler
```

```python
mvps = pd.read_csv('mvps.csv')

mvps
```

```python
mvps = mvps[["Player", "Year", "Pts Won", "Pts Max", "Share", "WS", "WS/48"]]

mvps.head()
```

## Cleaning the Players dataset

```python
players = pd.read_csv('players.csv')

players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
del players["Awards"]

players
```

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
players = players[~players["Player"].str.contains("League")]

players
```

```python
num_teams = list(range(2,6))

def single_row(df):
    if df.shape[0]==1:
        return df
    else:
        for num_team in num_teams:
            row = df[df["Team"] == "{}TM".format(num_team)]
            df["Team"] = df.iloc[-1,:]["Team"]
        return row
    
players = players.groupby(["Player", "Year"]).apply(single_row)
```

```python
players.head()
```

```python
print(players.keys())
print(mvps.keys())
```

```python
players.index = players.index.droplevel()
```

```python
players.index = players.index.droplevel()
```

```python
combined = players.merge(mvps, how="outer", on=["Player", "Year"])

combined
```

```python
combined[["Pts Won", "Pts Max", "Share"]] = combined[["Pts Won", "Pts Max", "Share"]].fillna(0)

combined
```

```python
teams = pd.read_csv("teams.csv")

teams.head()
```

```python
teams["Team"] = teams["Team"].str.replace("*", "", regex=False)

teams
```

```python
nicknames = {}

with open("nicknames.csv") as f:
    lines = f.readlines()
    for line in lines[1:]:
        abbrev, name = line.replace("\n", "").split(",")
        nicknames[abbrev] = name
```

```python
combined["Team"] = combined["Team"].map(nicknames)

combined
```

```python
stats = combined.merge(teams, how="left", on=["Team", "Year"])
```

```python
stats.shape
```

```python
del stats["Unnamed: 0"]
# del stats["GB"]

# rows_delete = list(range(14153,14160))
# stats.drop(rows_delete, inplace=True)
```

```python
# retain all players 
# stats = stats.fillna(0)
# remove players absent from mvps voting
stats = stats.dropna()


stats.to_csv("player_mvp_stats.csv")

stats
```

## Normalizing features before model input

```python
# Selecting features to scale
features = ['PTS', 'TRB', 'AST', 'STL', 'BLK', 'FG%', 'eFG%', 'WS', 'WS/48']
X = stats[features].copy()

# Fill NaNs with 0 (appropriate for counting stats where missing means zero value)
X = X.fillna(0)

# Initialize the scaler object
scaler = StandardScaler()

# Fit the scaler to the data (calculates μ and σ)
scaler.fit(X)

# Apply the transformation
X_scaled_array = scaler.transform(X)

# Convert back to a DataFrame for readability
X_scaled_df = pd.DataFrame(X_scaled_array, columns=[f'{col}_Scaled' for col in features])

# Ensure the index matches the original DataFrame for proper alignment
X_scaled_df = pd.DataFrame(
    X_scaled_array, 
    columns=[f'{col}_Scaled' for col in features],
    index=stats.index 
)

# combine with stats DF
merged_df = pd.concat([stats, X_scaled_df], axis=1)

# Save to CSV for manual inspection
merged_df.to_csv("player_mvp_stats_scaled.csv")
```

## Splitting Data into 3 sets: Training, Validation, Test

```python
# defining scaled features
scaled_features = ['PTS_Scaled', 'TRB_Scaled', 'AST_Scaled', 'STL_Scaled', 'BLK_Scaled', 'FG%_Scaled', 'eFG%_Scaled', 'WS_Scaled', 'WS/48_Scaled']


# Define year cutoffs
YEAR_TRAIN_END = 2018
YEAR_VALIDATION_END = 2021

# Training Set (1991-2018)
train_df = merged_df[merged_df['Year'] <= YEAR_TRAIN_END].copy()
x_train = train_df[scaled_features]
y_train = train_df['Share']

# Validation Set (2019-2021)
validation_df = merged_df[
    (merged_df['Year'] > YEAR_TRAIN_END) & 
    (merged_df['Year'] <= YEAR_VALIDATION_END)
].copy()
x_validation = validation_df[scaled_features]
y_validation = validation_df['Share']

# Test Set (2022-present)
test_df = merged_df[merged_df['Year'] > YEAR_VALIDATION_END].copy()
x_test = test_df[scaled_features]
y_test = test_df['Share']

print(f"Dataset years: \nTrain: {train_df['Year'].min()}-{train_df['Year'].max()} \nValidation: {validation_df['Year'].min()}-{validation_df['Year'].max()} \nTest: {test_df['Year'].min()}-{test_df['Year'].max()}")
```

## Model training

```python
from xgboost import XGBRegressor

# 1. Define the model
xgb_model = XGBRegressor(
    objective='reg:squarederror',
    n_estimators=1000,
    random_state=100
)

# 2. Fit the model to training data
xgb_model.fit(x_train, y_train)
```

```python
import matplotlib.pyplot as plt

# 1. Get feature importance dict
importance = xgb_model.feature_importances_

# 2. Create a DataFrame for visualization
feature_importance_df = pd.DataFrame({
    'Feature': features,
    'Importance': importance
}).sort_values(by='Importance', ascending=False)

# 3. Print the ranking
print("\n--- Feature Importance Ranking ---")
print(feature_importance_df)

# 4. Plot the results
plt.figure(figsize=(10, 6))
plt.barh(feature_importance_df['Feature'], feature_importance_df['Importance'], color='skyblue')
plt.xlabel('Relative Importance')
plt.title('Which Stats Determine the MVP?')
plt.gca().invert_yaxis() # Highest importance at the top
plt.show()
```

# Model Validation

```python
from sklearn.inspection import permutation_importance

# Calculate Permutation Importance on the validation set
perm_importance = permutation_importance(
    xgb_model, 
    x_validation, 
    y_validation, 
    n_repeats=10, 
    random_state=42,
    scoring='r2'
)

# Visualize
sorted_idx = perm_importance.importances_mean.argsort()

plt.figure(figsize=(10, 6))
plt.boxplot(
    perm_importance.importances[sorted_idx].T, 
    vert=False, 
    labels=x_validation.columns[sorted_idx]
)
plt.title("Permutation Importance (Test Set)")
plt.xlabel("Permutation Importance (Decrease in Score)")
plt.ylabel("Features")
plt.tight_layout()
plt.show()
```

## Making Predictions for Test Set: 2022-Now

```python
# Make predictions on the test set
predictions = xgb_model.predict(x_test)

# Create a DataFrame for comparison
results_df = test_df[['Player', 'Year', 'Share']].copy()
results_df['Predicted_Share'] = predictions

for year in results_df['Year'].unique():
    year_data = results_df[results_df['Year'] == year].copy()
    # create Actual Rank column
    year_data['Actual_Rank'] = year_data['Share'].rank(ascending=False, method='min')
    results_df.loc[year_data.index, 'Actual_Rank'] = year_data['Actual_Rank']

# Calculate the error
results_df['Error'] = results_df['Predicted_Share'] - results_df['Share']

# Function to show top candidates for a specific year
def show_top_candidates(year, df, n=10):
    year_data = df[df['Year'] == year].copy()
    # Sort by Predicted Share to see who the model thought would win
    year_data = year_data.sort_values(by='Predicted_Share', ascending=False)
    
    # Add a Rank column based on prediction
    year_data['Predicted_Rank'] = range(1, len(year_data) + 1)
    
    # Add Actual Rank column
    # year_data['Actual_Rank'] = year_data['Share'].rank(ascending=False, method='min')
    
    return year_data[['Player', 'Predicted_Share', 'Share', 'Predicted_Rank', 'Actual_Rank', 'Error']].head(n)

# Display top predictions for each year in the test set (2022, 2023, 2024, 2025)
years_in_test = sorted(results_df['Year'].unique())

for year in years_in_test:
    print(f"\n--- Top Predicted MVP Candidates for {year} ---")
    print(show_top_candidates(year, results_df))

# Calculate Spearman Rank Correlation for each year to measure ranking accuracy
from scipy.stats import spearmanr

print("\n--- Model Accuracy (Spearman Rank Correlation) ---")
for year in years_in_test:
    year_data = results_df[results_df['Year'] == year]
    # We only care about the correlation for the top candidates, but calculating on all gives a general sense.
    # Often, filtering for players with >0 actual or predicted votes helps reduce noise from the 400+ non-candidates.
    # Let's look at the correlation for the whole set first.
    corr, _ = spearmanr(year_data['Predicted_Share'], year_data['Share'])
    print(f"{year}: {corr:.4f}")
```

## Visualizing Predictions

```python
import seaborn as sns
def visualize_mvp_predictions(df):
    """
    Visualizes MVP predictions vs actual results.
    Assumes df has columns: 'Player', 'Year', 'Predicted_Share', 'Share', 'Actual_Rank'
    """
    years = sorted(df['Year'].unique())
    
    # Setup the plot
    fig, axes = plt.subplots(len(years), 1, figsize=(12, 5 * len(years)), sharex=False)
    if len(years) == 1:
        axes = [axes]
    
    # Custom color palette
    colors = sns.color_palette("muted")
    
    for i, year in enumerate(years):
        ax = axes[i]
        
        # Filter top 8 predicted players for the year
        year_data = df[df['Year'] == year].sort_values('Predicted_Share', ascending=False).head(8)
        
        # Data for plotting
        players = year_data['Player']
        x = np.arange(len(players))
        width = 0.35
        
        # Bars
        rects1 = ax.bar(x - width/2, year_data['Predicted_Share'], width, label='Predicted Share', color='skyblue', alpha=0.9)
        rects2 = ax.bar(x + width/2, year_data['Share'], width, label='Actual Share', color='lightgray', alpha=0.9)
        
        # Highlight winner bars (Actual Rank 1)
        # Find the index of the actual winner in this sorted list
        winner_mask = year_data['Actual_Rank'] == 1.0
        if winner_mask.any():
            winner_idx = np.where(winner_mask)[0][0]
            # Highlight the Actual Share bar for the winner
            rects2[winner_idx].set_color('#FFD700') # Gold
            rects2[winner_idx].set_edgecolor('black')
            rects2[winner_idx].set_linewidth(1.5)
            # Add a "Winner" annotation
            ax.annotate('Winner', 
                        xy=(x[winner_idx] + width/2, year_data.iloc[winner_idx]['Share']), 
                        xytext=(0, 2), textcoords='offset points',
                        ha='center', va='bottom', fontsize=10, fontweight='bold', color='black')

        # Formatting
        ax.set_ylabel('Vote Share')
        ax.set_title(f'{year} MVP Predictions: Model vs. Reality', fontsize=14, fontweight='bold')
        ax.set_xticks(x)
        ax.set_xticklabels(players, rotation=15, ha='right')
        ax.legend()
        ax.grid(axis='y', linestyle=':', alpha=0.4)
        
    plt.tight_layout()
    # rename for mvp only set
    plt.savefig('mvp_predictions_viz.png')
    plt.show()

visualize_mvp_predictions(results_df)
```
