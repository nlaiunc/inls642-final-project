## Cleaning the MVPs dataset

```python
mvps = mvps[["Player", "WS", "WS/48", "Year", "Pts Won", "Pts Max", "Share"]]

mvps.head()
```

## Cleaning the Players dataset

```python
players = pd.read_csv('players.csv')

players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
del players["Awards"]

players
```

Merge MVP and Players dataset

```python
players = players.join(mvps.set_index(['Player', 'Year']), on=['Player', 'Year'], how='left')
players
```

Cleaning teams dataset

```python
teams = pd.read_csv('teams.csv')

print(teams.head())
```

```python
del teams["Unnamed: 0"]
# clean 'GB' column, convert dashes to 0. 
teams['GB'] = teams['GB'].astype(str)
teams['GB'] = teams['GB'].replace(r'^\s*(—|–|--|-)\s*$', '0', regex=True)
teams['GB'] = pd.to_numeric(teams['GB'], errors='coerce').fillna(0.0).astype(float)

teams
```
